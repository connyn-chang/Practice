```python
import random
```

```python
my = int(input("0.剪刀 1.石頭 2.布:"))
com = random.randint(0, 2)
trans = ["剪刀", "石頭", "布"]
print("我：", trans[my])
print("電腦：", trans[com])

if my == (com + 1) % 3:
    print("WIN :)")
elif com == (my + 1) % 3:
    print("Lose :(")
else:
    print("Even")
```

```python
my1 = int(input("0.棒 1.老虎 2.雞 3.蟲"))
com1 = random.randint(0, 3)
trans1 = ["棒", "老虎", "雞", "蟲"]
print("我:", trans1[my1])
print("電腦:", trans1[com1])

if my1 == (com1 - 1) % 4:
    print("I Win :)")

elif com1 == (my1 - 1) % 4:
    print("I Lose :(")

else:
    print("Even")
```

```python
#改進1
layer = 10
i = 0
while i < layer:
    lcount = layer - i - 1
    rcount = i + 1
    print("" * lcount + "*" * rcount)
    i = i + 1
```

```python
#第i行需要印的星星數= 2i - 1
#第i行需要印的空格數= 總行數 - i

left, middle, right = " ", "*", " "
layer = 10
i = 0
while i < layer:
  mcount = 2 * i - 1
  lcount = layer - i
  rcount = lcount
  print(left * lcount + middle * mcount + right * rcount)
  i = i + 1
```
