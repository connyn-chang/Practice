```python
import random

diameter = int(input("請輸入直徑"))
incircle, outcircle = 0, 0
i = 0
while i < 1000000:
    x = random.uniform(-diameter/2, diameter/2)
    y = random.uniform(-diameter/2, diameter/2)

    if x ** 2 + y ** 2 < 1:
        #print("in")
        incircle = incircle + 1
    else:
        #print(out)
        outcircle = outcircle + 1
    i = i + 1

print(incircle, outcircle)
totalarea = diameter ** 2
ratio = incircle / (incircle + outcircle)
print("面積是：", totalarea * ratio)
```
